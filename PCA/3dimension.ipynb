```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
x = np.array([1,2,4])
y = np.array([3,5,7])
z = np.array([2,1,3])
```

```python
ax.scatter3D(x,y,z, alpha=0.6)
```

```python
for x, y, z in zip(x, y, z):
    label = '(%d, %d, %d)' % (x, y, z)
    ax.text(x, y, z, label)
```
