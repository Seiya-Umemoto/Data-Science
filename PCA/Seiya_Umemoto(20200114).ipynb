```python
data = pd.read_csv('MNIST_100.csv')
X = data.drop('label', axis=1)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
PCAX = pca.transform(X)

plt.figure(figsize=(10,10))

for i in range(0, 901, 100):
    plt.scatter(PCAX[i:i+100, 0], PCAX[i:i+100, 1], s=100, marker='$'+str(i//100)+'$', alpha=0.7)
plt.title("2-dimensions PCA")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

pca = PCA(n_components=3)
pca.fit(X)
PCAX = pca.transform(X)

fig = plt.figure(figsize=(7,6))
ax = fig.add_subplot(projection='3d')

for i in range(0, 901, 100):
    ax.scatter(PCAX[i:i+100, 0], PCAX[i:i+100, 1], PCAX[i:i+100, 2], s=15, marker='$'+str(i//100)+'$', alpha=0.7)
ax.set_title("3 dimensions PCA")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.show()
```
